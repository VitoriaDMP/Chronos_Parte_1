```sql
CREATE TABLE DISEASE (
    id_disease INTEGER NOT NULL AUTO_INCREMENT, 
    name VARCHAR(50) NOT NULL, 
    category VARCHAR(30), 
    PRIMARY KEY(id_disease)
);

CREATE TABLE DIAGNOSIS(
    id_diagnosis INTEGER NOT NULL,
    id_disease INTEGER NOT NULL,
    PRIMARY KEY(id_diagnosis),
    FOREIGN KEY(id_disease) 
        REFERENCES DISEASE(id_disease)
        ON DELETE NO ACTION
        ON UPDATE NO ACTION
);

CREATE TABLE EXAM(
    id_exam INTEGER NOT NULL,
    qtd_itens INTEGER,
    PRIMARY KEY(id_exam)
);

CREATE TABLE ITEM(
    id_item INTEGER NOT NULL,
    id_exam INTEGER NOT NULL,
    name VARCHAR(50) NOT NULL,
    valor INTEGER NOT NULL,
    PRIMARY KEY(id_item),
    FOREIGN KEY(id_exam) 
        REFERENCES EXAM(id_exam)
        ON DELETE NO ACTION
        ON UPDATE NO ACTION
    
);

CREATE TABLE PATIENT(
    id_patient INTEGER NOT NULL,
    id_diagnosis INTEGER NULL, 
    id_exam INTEGER NOT NULL,
    age INTEGER, 
    sex CHAR, 
    pregnant BOOLEAN, 
    tumor BOOLEAN, 
    sick BOOLEAN, 
    psych BOOLEAN, 
    goitre BOOLEAN,
    PRIMARY KEY(id_patient),
    FOREIGN KEY(id_diagnosis) 
        REFERENCES DIAGNOSIS(id_diagnosis)
        ON DELETE NO ACTION
        ON UPDATE NO ACTION,
    FOREIGN KEY(id_exam) 
        REFERENCES EXAM(id_exam)
        ON DELETE NO ACTION
        ON UPDATE NO ACTION,
    );
```
